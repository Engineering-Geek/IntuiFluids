```python
from sympy import symbols, Eq
from meshing.mesh import generate_boundary_from_equation, BoundarySurface
import numpy as np
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
x, y, z = symbols("x y z")
a = 1
b = 1
c = 0.5
d = 1
equation = Eq(x**2 / a**2 + y**2 / b**2 - z**2 / c**2, d)
```

```python
boundary = generate_boundary_from_equation(
    equation,
    "dirichlet",
    x_values=np.linspace(-20, 20, 101),
    z_values=np.linspace(0, 5, 51),
)
# plot the coordinates
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
coordinates = [node.coordinates for node in boundary.nodes]
x = [coordinate[0] for coordinate in coordinates]
y = [coordinate[1] for coordinate in coordinates]
z = [coordinate[2] for coordinate in coordinates]
ax.scatter(x, y, z)
ax.set_title("Boundary coordinates")
# label the axes
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
plt.show()
```
